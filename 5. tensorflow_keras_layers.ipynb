```python
import numpy as np
import tensorflow as tf
import logging
logging.getLogger("tensorflow").setLevel(logging.ERROR)

```

```python
X_train = np.array([1.0, 2.0]).reshape(-1, 1)
Y_train = np.array([300.0, 500.0]).reshape(-1, 1)
print(X_train)
print(Y_train)
```

```python
linear_layer = tf.keras.layers.Dense(units=1, activation='linear')
```

```python
linear_layer.get_weights()
```

```python
a1 = linear_layer(X_train[0].reshape(1, 1))
w, b = linear_layer.get_weights()
print(w, b)
```

```python
set_w = np.array([[200]])
set_b = np.array([100])

linear_layer.set_weights([set_w, set_b])
print(linear_layer.get_weights())
```

```python
a1 = linear_layer(X_train[0].reshape(-1, 1))
print(a1)
alin = X_train[0].reshape(-1, 1) @ set_w + set_b
print(alin)
```

```python
pred_tf = linear_layer(tf.constant(X_train))
pred_np = X_train @ set_w + set_b

print(pred_tf)
print(pred_np)
```

```python
X_train = np.array([0., 1, 2, 3, 4, 5], dtype=np.float32).reshape(-1,1)  # 2-D Matrix
Y_train = np.array([0,  0, 0, 1, 1, 1], dtype=np.float32).reshape(-1,1)  # 2-D Matrix
```

```python
model = tf.keras.Sequential([
    tf.keras.Input(shape=(1,)),
    tf.keras.layers.Dense(1, activation='sigmoid', name='L1')
])
model.summary()
```

```python
logistic_layer = model.get_layer('L1')
```

```python
w, b = logistic_layer.get_weights()
print(w, b)
print(w.shape, b.shape)
```

```python
set_w = np.array([[2]])
set_b = np.array([-4.5])
logistic_layer.set_weights([set_w, set_b])
print(logistic_layer.get_weights())
```

```python
def signmoid(e):
    return 1 / (1 + np.exp(-e))
```

```python
a1 = model.predict(X_train[0].reshape(-1, 1))
alog = signmoid(set_w @ X_train[0].reshape(-1, 1) + set_b)
print(a1, alog)
```
