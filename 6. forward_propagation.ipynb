```python
# forward prop
import numpy as np
```

$$a_1^{[1]} = g(\vec{w}_1^{[1]} \cdot \vec{x} + b_1^{[1]})$$

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
x = np.array([200, 17])

w1_1 = np.array([1, 2])
b1_1 = np.array([-1])
z1_1 = w1_1 @ x + b1_1
a1_1 = sigmoid(z1_1)

w1_2 = np.array([-3, 4])
b1_2 = np.array([1])
z1_2 = w1_2 @ x + b1_2
a1_2 = sigmoid(z1_2)

w1_3 = np.array([5, -6])
b1_3 = np.array([1])
z1_3 = w1_3 @ x + b1_3
a1_3 = sigmoid(z1_3)

a1 = np.array([a1_1, a1_2, a1_3])

w2_1 = np.array([-7, 8, 9])
b2_1 = np.array([3])
z2_1 = w2_1 @ a1 + b2_1
a2_1 = sigmoid(z2_1)

print(a2_1)
```

```python
g = sigmoid
def dense(a_in, W, b):
    units = W.shape[1]
    a_out = np.zeros(units)
    for j in range(units):
        w = W[:, j]
        z = w @ a_in + b[j]
        a_out[j] = g(z)
    
    return a_out
```

```python
W1 = np.stack((w1_1, w1_2, w1_3), axis=-1)
b1 = np.concat((b1_1, b1_2, b1_3))
d = dense(x, W1, b1)
print(d)
```

```python
X = np.array([[200, 17]])
B = np.array([[-1, 1, 2]])

def dense_vectorization(A_in, W, B):
    Z = A_in @ W + B
    A_out = g(Z)
    return A_out

print(dense_vectorization(X, W1, B))
```
