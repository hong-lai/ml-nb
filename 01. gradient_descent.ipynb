```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_train = np.array([1., 2.])
y_train = np.array([300., 500.])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
w = 200
b = 100
```

```python
from numpy import ndarray


def compute_model_ouptut(x: ndarray, w, b):
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
    return f_wb
```

```python
tmp_f_wb = compute_model_ouptut(x_train, w, b)
plt.plot(x_train, tmp_f_wb, c='b', label='Our Prediction')
plt.scatter(x_train, y_train, marker='x', c='r', label='Actual Values')
plt.title('Housing Prices')
plt.ylabel('Price (in 1000s of dollars)')
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```

```python
def compute_cost(x_train, y_train, w, b):
    m = x_train.shape[0]
    cost = 0
    for x, y in zip(x_train, y_train):
        y_hat = w * x + b
        cost += (y_hat - y) ** 2

    return 1 / (2 * m) * cost
```

```python
# 1. Define the range for the weight parameter w
J_x = np.arange(150, 251)

# 2. Compute the cost for each w while holding b constant at 100
J_y = np.array([compute_cost(x_train, y_train, i, 100) for i in J_x])

# 3. Calculate the cost at the specific point of interest (w=200, b=100)
point_cost = compute_cost(x_train, y_train, 200, 100)
print(f"Cost at (w=200, b=100): {point_cost}")

# 4. Plot the cost curve line
plt.plot(J_x, J_y, c='c', label='$J(w, b=100)$')

# 5. Correctly plot the single cost point on the curve
plt.scatter(200, point_cost, c='b', marker='o', s=100, zorder=5, label='Evaluated Point')

# 6. Add labels and title
plt.xlabel('Weight (w)')
plt.ylabel('Cost J(w, b)')
plt.title("Cost function $J(w, b)$ holding $b=100$")
plt.legend()
plt.show()
```

```python

def compute_gradient_descent(x, y):
    num_iters = 1000
    m = x.shape[0]
    alpha = 0.4
    w = 0
    b = 0
    for _ in range(num_iters):
        dj_dw = np.mean([(w * x[i] + b - y[i]) * x[i]
                         for i in range(m)])
        dj_db = np.mean([w * x[i] + b - y[i]
                         for i in range(m)])

        w = w - alpha * dj_dw
        b = b - alpha * dj_db

    return (w, b)
```

```python
w_final, b_final = compute_gradient_descent(x_train, y_train)
print(f"(w, b) found by Gradient Descent: ({w_final:.5f}, {b_final:.5f})")
```
